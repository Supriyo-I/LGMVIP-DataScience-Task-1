```python
columns=['sepal length','sepal width','petal length','peta width','class_lables']
df=pd.read_csv('/iris.data',names=columns)
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df,hue='class_lables')
```

Seperating input columns and output columns

```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

Using support vector machine algorithum

```python
from sklearn.svm import SVC

node1_svc=SVC()
node1_svc.fit(X_train,y_train)
```

```python
prediction1=node1_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
```

Using logistic regression

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
predect=model_LR.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predect)*100)
```

Detailed Classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction1))
```

Taking Data input
